# **YoloV3**

## **Loading the model:**

```python
modelv3 = YOLO('yolov3u.pt')
```

## **Using the model:**

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
vid = '/content/gdrive/MyDrive/Test_sample_video.mp4'
```

```python
result = modelv3.predict(vid)[0]
```

## **Getting the output video:**

```python
output_path = '/content/gdrive/MyDrive/Yolo_output/vid_v3.mp4'


cap = cv2.VideoCapture(vid)
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))


fourcc = cv2.VideoWriter_fourcc(*'mp4v')
encoder = cv2.VideoWriter(output_path, fourcc, 25, (frame_width, frame_height))

model = YOLO('yolov3u.pt')

time_taken = 0
total_detected_objects = 0

while 1 :
    ret, frame = cap.read()
    if not ret :
      break

    result = model.predict(frame)[0]

    im = cv2.cvtColor(result.plot(), cv2.COLOR_BGR2RGB)  # get the labeled image
    encoder.write(im)

    time_taken +=  sum(result.speed.values())
    total_detected_objects += len(result.boxes.data)


cap.release()
encoder.release()
print(f'Total detected {total_detected_objects}, Total time {time_taken :.2f}')
```

## **Printing Number of objects:**

```python
print(f'Number of objects in the image', total_detected_objects)
```

## **Getting the time (CPU):**

```python
print(f'Time : {result.speed}')
print(f'Total time (CPU) : {time_taken}')
```
